# Modelling

## Import Relevant Libraries

```python
import scipy.stats as stats

from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn import metrics
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, accuracy_score, plot_roc_curve, roc_auc_score, recall_score, precision_score, f1_score

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

### Preprocess our input data first

```python
train.head()
```

```python
# features = [col for col in train._get_numeric_data().columns if col != 'survived']
features = ['sex', 'sibsp', 'parch', 'pclass']
py_features = ['sex', 'sibsp', 'parch', 'pclass', 'survived']

X = train[features]
Y = train['survived']

pycaret_data = train[py_features]
```

```python
X.head()
```

### Train Test Split our Training Data First

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=100)
```

```python
X_train.head()
```

## Scale Split Training Data

```python
ss = StandardScaler()
ss.fit(X_train)
X_train = ss.transform(X_train)
X_test = ss.transform(X_test)
```

## Creating a Function for our Models

```python
def model_scores(model, x_train, y_train, x_test, y_test):
    cv_scores = (cross_val_score(model, x_train, y_train)).mean()
    
    model.fit(x_train, y_train)
    predictions = model.predict(x_train)
    
    rmse = metrics.mean_squared_error(y_train, predictions)
    mae = metrics.mean_absolute_error(y_train, predictions)
    rsquare = metrics.r2_score(y_train, predictions)
    
    print("Cross-Val = " + str(cv_scores))
    print("Train RMSE = " + str(rmse))
    print("Train MAE = " + str(mae))
    print("Train R2 Score = " + str(rsquare))
    print("--------")
    
    test_pred = model.predict(x_test)
    test_rmse = metrics.mean_squared_error(y_test, test_pred)
    test_mae = metrics.mean_absolute_error(y_test, test_pred)
    test_rsquare = metrics.r2_score(y_test, test_pred) 
    
    print("Test RMSE = " + str(test_rmse))
    print("Test MAE = " + str(test_mae))
    print("Test R2 Score = " + str(test_rsquare))
    print("--------")
#     print(classification_report(y_test, test_pred))
    
#     confusion_matrix(y_test, test_pred)
    
#     plot_roc_curve(model, X_test, y_test)
#     plt.plot([0, 1], [0, 1],label='baseline', linestyle='--')
#     plt.legend();
    
    return None
```

```python
model_scores(LinearRegression(),X_train, y_train, X_test, y_test)
```

```python
model_scores(LassoCV(n_alphas = 400),X_train, y_train, X_test, y_test)
```

```python
model_scores(LogisticRegression(),X_train, y_train, X_test, y_test)
```

```python
model_scores(DecisionTreeClassifier(),X_train, y_train, X_test, y_test)
```
